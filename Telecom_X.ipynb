#üìå Extracci√≥n

```python
# Importamos las librer√≠as necesarias para trabajar con datos
import requests
import pandas as pd

# Definimos la direcci√≥n del archivo JSON con datos de clientes
enlace_datos = "https://raw.githubusercontent.com/adsecv/ctx_aec_1/refs/heads/main/Base-de-datos/TelecomX_Data.json"

# Obtenemos los datos desde la URL
respuesta = requests.get(enlace_datos)

# Convertimos los datos JSON en un objeto de Python
datos_json = respuesta.json()

# Creamos una tabla con los datos obtenidos
tabla_clientes = pd.DataFrame(datos_json)

# Visualizamos la tabla para explorar los datos
tabla_clientes
```

```python
# Aplanamos los datos JSON para trabajar con una tabla m√°s simple
tabla_normalizada = pd.json_normalize(datos_json)

# Mostramos la tabla normalizada para verificar su contenido
tabla_normalizada
```

```python
# Visualizamos la estructura y tipos de datos de la tabla normalizada
tabla_normalizada.info()
```

```python
# Exploramos los valores √∫nicos de cada columna en la tabla
for columna in tabla_normalizada.columns:
    cantidad_unicos = tabla_normalizada[columna].nunique()
    print(f"Columna '{columna}' tiene {cantidad_unicos} valores √∫nicos")
    if cantidad_unicos < 20:  # Mostramos valores solo si son pocos
        valores = tabla_normalizada[columna].unique()
        print(f"Valores: {valores}")
        print("=" * 60)  # Separador para claridad en la salida
```

```python
# Comprobamos si existen filas duplicadas en la tabla
total_duplicados = tabla_normalizada.duplicated().sum()
print(f"Se encontraron {total_duplicados} filas duplicadas en el dataset.")
```

```python
# Revisamos la presencia de valores nulos en la tabla
nulos_por_columna = tabla_normalizada.isnull().sum()
print(f"Cantidad de valores nulos por columna:\n{nulos_por_columna}")
```

```python
# Comprobamos si hay celdas con cadenas vac√≠as o espacios en blanco
cadenas_vacias = tabla_normalizada.apply(lambda col: col.astype(str).str.strip() == '').sum()
print(f"Cantidad de valores vac√≠os o en blanco por columna:\n{cadenas_vacias}")
```

#üîß Transformaci√≥n

```python
# Convertimos las columnas de cargos totales y mensuales a formato num√©rico

tabla_normalizada["account.Charges.Total"] = pd.to_numeric(tabla_normalizada["account.Charges.Total"], errors="coerce")
tabla_normalizada["account.Charges.Monthly"] = pd.to_numeric(tabla_normalizada["account.Charges.Monthly"], errors="coerce")

print("Tipo de dato (Total):", tabla_normalizada["account.Charges.Total"].dtype)
print("Tipo de dato (Mensual):", tabla_normalizada["account.Charges.Monthly"].dtype)
```

```python
# Eliminamos las filas con valores vac√≠os o espacios en blanco en la columna 'Churn'

tabla_normalizada = tabla_normalizada[tabla_normalizada['Churn'].str.strip() != '']
print(f"Filas restantes despu√©s de eliminar las vac√≠as en 'Churn': {len(tabla_normalizada)}")
```

```python
# Contamos las celdas vac√≠as o con solo espacios en blanco

tabla_normalizada.apply(lambda x: x.astype(str).str.strip() == '').sum()
```

```python
# Generamos un resumen estad√≠stico de las columnas num√©ricas

tabla_normalizada.describe()
```

```python
# Calculamos la cantidad diaria de cargos a partir del monto mensual

tabla_normalizada.loc[:, 'Cuentas_Diarias'] = tabla_normalizada['account.Charges.Monthly'] / 30
print(tabla_normalizada[['account.Charges.Monthly', 'Cuentas_Diarias']])
```

```python
# Eliminamos la fila 'Cuentas_Diarias'

tabla_normalizada = tabla_normalizada.drop('Cuentas_Diarias', errors='ignore')
```

```python
# Importamos la librer√≠a Plotly para visualizaci√≥n
import plotly.express as px

# Creamos un histograma para explorar la relaci√≥n entre g√©nero y Churn
px.histogram(tabla_normalizada, x='customer.gender', text_auto=True, color='Churn', barmode='group')
```

```python
# Agrupamos los datos por 'tenure' y estado de 'Churn'
grupo_tenure_churn = tabla_normalizada.groupby(['customer.tenure', 'Churn']).size().unstack(fill_value=0)

# Calculamos el porcentaje de churn para cada periodo de tenure
grupo_tenure_churn['tasa_churn'] = (grupo_tenure_churn.get('Yes', 0) / grupo_tenure_churn.sum(axis=1)) * 100

# Imprimimos la tasa de churn por tenure
print(grupo_tenure_churn[['tasa_churn']])
```

```python
# Graficamos la evoluci√≥n de la tasa de churn en funci√≥n del tenure
grafico_churn = px.line(grupo_tenure_churn.reset_index(),
                       x='customer.tenure',
                       y='tasa_churn',
                       markers=True,
                       title='Porcentaje de Churn seg√∫n Tenure')
grafico_churn.show()
```

```python
# An√°lisis de correlaci√≥n con variables clave
import seaborn as sns
import matplotlib.pyplot as plt

# Columnas de servicios a convertir a formato binario
columnas_servicios = [
    'phone.PhoneService',
    'phone.MultipleLines',
    'internet.InternetService',
    'internet.OnlineSecurity',
    'internet.OnlineBackup',
    'internet.DeviceProtection',
    'internet.TechSupport',
    'internet.StreamingTV',
    'internet.StreamingMovies'
]

# Diccionario de valores que cuentan como servicio activo por columna
servicio_activo = {
    'phone.PhoneService': ['Yes'],
    'phone.MultipleLines': ['Yes'],
    'internet.InternetService': ['DSL', 'Fiber optic'],
    'internet.OnlineSecurity': ['Yes'],
    'internet.OnlineBackup': ['Yes'],
    'internet.DeviceProtection': ['Yes'],
    'internet.TechSupport': ['Yes'],
    'internet.StreamingTV': ['Yes'],
    'internet.StreamingMovies': ['Yes']
}

# Nuevo DataFrame para el an√°lisis de correlaci√≥n
df_correlacion = pd.DataFrame()

# Agregamos la columna binaria de churn
df_correlacion['Churn_bin'] = tabla_normalizada['Churn'].apply(
    lambda x: 1 if str(x).strip().lower() == 'yes' else 0
)

# Agregamos la columna ya existente 'Cuentas_Diarias'
df_correlacion['Cuentas_Diarias'] = tabla_normalizada['Cuentas_Diarias']

# Convertimos los servicios a binario seg√∫n el diccionario
for col in columnas_servicios:
    nueva_col = col + '_bin'
    df_correlacion[nueva_col] = tabla_normalizada[col].apply(
        lambda x: 1 if str(x).strip() in servicio_activo[col] else 0
    )

# Agregamos la cantidad total de servicios contratados
servicios_bin_cols = [col + '_bin' for col in columnas_servicios]
df_correlacion['Cantidad_Servicios'] = df_correlacion[servicios_bin_cols].sum(axis=1)

# Calculamos la matriz de correlaci√≥n
columnas_analisis = ['Churn_bin', 'Cuentas_Diarias', 'Cantidad_Servicios'] + servicios_bin_cols
matriz_corr = df_correlacion[columnas_analisis].corr()

# Mostramos la matriz de correlaci√≥n
print("Matriz de correlaci√≥n entre variables seleccionadas y Churn:")
print(matriz_corr)

# Visualizaci√≥n con heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(matriz_corr, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title('Matriz de Correlaci√≥n: Servicios, Cuentas Diarias y Churn')
plt.tight_layout()
plt.show()
```

- Mayor√≠a de abandonos ocurre antes de 10 meses.

Conclusi√≥n: Los primeros 6 meses son clave (churn >40%).

## Correlaciones
- **Con Churn**:
  - Cargos diarios: +0.39 (m√°s cargos, m√°s churn).
  - Servicios: -0.25 (m√°s servicios, menos churn).
  - Internet: +0.31 (fibra √≥ptica 42%, DSL 19%, sin internet 7%).
  - Seguridad y soporte t√©cnico: -0.17 y -0.16 (reducen churn).
- Clientes que abandonan tienen cargos m√°s altos (mediana $80 vs. $65).
- Mayor churn con tenure bajo (<10 meses) y cargos altos (>$80).
- Servicio telef√≥nico no afecta churn (~26%).
- StreamingTV: 30% churn con servicio, 25% sin √©l.

**Observaciones**: Cargos altos y fibra √≥ptica aumentan churn; seguridad y soporte t√©cnico retienen.

## Hecho Interesante
62% de clientes con 2 meses de permanencia abandonan, muy por encima del promedio (26.5%).

## Conclusiones
1. Churn cr√≠tico en primeros 6 meses (pico 62% a 2 meses).
2. Contratos mes a mes (43%) y cheques electr√≥nicos (45%) son riesgos.
3. Fibra √≥ptica (42%) y cargos altos fomentan abandono.
4. M√°s servicios (seguridad, soporte) reducen churn.

## Estrategias
- Descuentos y soporte en primeros 6 meses.
- Promover contratos de 1‚Äì2 a√±os.
- Revisar precios y calidad de fibra √≥ptica.
